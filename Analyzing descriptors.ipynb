```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = range(80)
y = np.zeros(80)

for i in range(80):
    file_name = 'Descriptors_compressed/descriptors_reduced'
    file_name += str(i)
    file_name += '.csv'
    try:
        df = pd.read_csv(file_name)
    except FileNotFoundError:
        print('File does not exist')
    else:
        df = pd.read_csv(file_name)
        y[i] = df.shape[0]
```

```python
plt.plot(x,y)
plt.xlabel("Image number through the first class")
plt.ylabel("Quantity of descriptors")
```
